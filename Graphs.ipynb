```python
import matplotlib.pyplot as plt


categories = ['Initial', 'VCloak', 'VCloak+RIR']
values = [3.8, 14.5, 29]

fig, ax = plt.subplots(figsize=(8, 6))

bars = ax.bar(categories, values, color='skyblue')
for bar, value in zip(bars, values):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01,
            f'{value:.2f}', ha='center', va='bottom', color='gray', fontsize=15)
ax.set_xlabel('Categories', fontsize=16)
ax.set_ylabel('WER (%)', fontsize=16)
ax.set_title('ASR model performance on anonimized data', fontsize=18)
ax.set_ylim(0, 100)
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=12)
ax.grid(True, axis='y', linestyle='--', alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('Picture_2.png', format='png')
plt.show()
```

```python
models = ['ECAPA', 'XVectorSincNet', 'Wespeaker']
scores = [
    [0.99, 0.12, 0.05], 
    [0.97, 0.07, 0.02], 
    [0.99, 0.44, 0.26]
]

fig, ax = plt.subplots(figsize=(12, 8))

bar_width = 0.25

r1 = range(len(scores))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.bar(r1, [score[0] for score in scores], color='blue', width=bar_width, edgecolor='grey', label='Score 1')
plt.bar(r2, [score[1] for score in scores], color='orange', width=bar_width, edgecolor='grey', label='Score 2')
plt.bar(r3, [score[2] for score in scores], color='green', width=bar_width, edgecolor='grey', label='Score 3')

plt.xlabel('Models', fontweight='bold')
plt.ylabel('Scores')
plt.title('Scores for Different Models')
plt.xticks([r + bar_width for r in range(len(scores))], models)

plt.legend()

plt.show()
```

```python
scores = {
    'Initial': [0.99, 0.99, 0.97],
    'VCloak': [0.44, 0.12, 0.07],
    'VCloak+RIR': [0.26, 0.05, 0.02]
}
models = ['Wespeaker', 'ECAPA', 'XVectorSincNet']

colors = ['skyblue', 'lightgreen', 'salmon']
# colors = ['plum', 'lightgreen', 'lemonchiffon']

fig, ax = plt.subplots(figsize=(9, 6))

bar_width = 0.3

r1 = range(len(scores))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.bar(r1, [scores['Initial'][0], scores['VCloak'][0], scores['VCloak+RIR'][0]], color=colors[0], width=bar_width, edgecolor='grey', label=models[0])
plt.bar(r2, [scores['Initial'][1], scores['VCloak'][1], scores['VCloak+RIR'][1]], color=colors[1], width=bar_width, edgecolor='grey', label=models[1])
plt.bar(r3, [scores['Initial'][2], scores['VCloak'][2], scores['VCloak+RIR'][2]], color=colors[2], width=bar_width, edgecolor='grey', label=models[2])

plt.xlabel('Data', fontweight='bold', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
plt.title('Speaker verification task by model', fontsize=18)
plt.xticks([r + bar_width for r in range(len(scores))], list(scores.keys()))

plt.grid(True, axis='y', linestyle='--', alpha=0.7)

ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=12)

plt.legend(title='Models', fontsize=13)

plt.tight_layout()
plt.savefig('Picture_1.png', format='png')
plt.show()
```
